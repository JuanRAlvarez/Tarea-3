```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure(figsize(9.5,9.5))
ax = Axes3D(fig)
n_points=5000

A =random.random(n_points)
B = random.random(n_points)
C = 2*B-1
R = 50
r = R*(random.random(n_points))**(1/3.0)
theta = 2*pi*A
phi = arccos(C)

x = r * cos(theta) * sin(phi)
y = r * sin(theta) * sin(phi)
z = r * cos(phi)

ax.scatter(x,y,z)
```
